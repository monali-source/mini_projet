Vu que cette donnée ne vient pas de notre table, on entre les données manuellement

```python
dicArtists = {
    'Katy Perry':"Santa Barbara",
    'Justin Bieber':"London Canada",
     'Rihanna':"Saint Michael",
    'Maroon 5':"Los Angeles",
    'Lady Gaga':"Manhattan",
    'Bruno Mars':"Honolulu", 
    'The Chainsmokers':"Times Square" ,
    'Pitbull':"Miami",
    'Shawn Mendes':"Toronto",
    'Ed Sheeran':"United Kingdom", 
  }
```

définit les coordonnées en longitude et lattitude

```python
import geocoder
listGeo = []

for value in (dicArtists.values()):
    g = geocoder.arcgis(value)
    listGeo.append(g.latlng)
```

```python
top_genres =[]
for key in (dicArtists.keys()):
    top_genres.append(data[data['artist']== key].top_genre.unique())
```

```python
lat = []
log = []
for i in listGeo:
    lat.append(i[0])
    log.append(i[1])
```

```python
colors = {
 'dance pop': 'pink',
 'pop': 'blue',
 'barbadian pop': 'green',
 'electropop': 'orange',
 'canadian pop': 'red',
}
```

```python
#j'ai eu une erreur mais je n'ai pas réussi à me régaler (line 2)

dfLocation = pd.DataFrame(columns=['Name', 'Lat', 'Log', 'Gen'])
dfLocation['Name'] = list(dicArtists.keys())
dfLocation['Gen'] = np.array(top_genres)
dfLocation['Lat'] = lat
dfLocation['Log'] = log

print(dfLocation)
```

Omodélise une carte via folium (qui utilise Google Maps) pour représenter les zones correspondantes

```python
spotify = folium.Map(
    location=[41.5503200,-8.4200500],
    zoom_start=2
)
spotify
```

Enfin,  affiche les artistes dans leur coordonnées sur cette carte

```python
for i in range(10):
    singer = dfLocation.iloc[i]
    folium.Marker(
        
        popup=singer['Name']+'-'+singer['Gen'],
        location=[singer['Lat'], singer['Log']],
    icon=folium.Icon(color=colors[singer['Gen']], icon='music')).add_to(spotify)
    
spotify
```
